## 3. 📦 Parameters & Arguments

- **Parameter** = variable in the function definition (placeholder)
- **Argument** = actual value passed when calling the function

```
def greet(name):      ← 'name' is a PARAMETER
    print(f"Hi {name}")

greet("Alice")        ← "Alice" is an ARGUMENT
```

````python
def greet(name):
    print(f"Hello, {name}! Nice to meet you! 🤝")

greet("Alice")
greet("Bob")
greet("Charlie")
````

````python
# 📌 Example 2: Multiple parameters
def describe_pet(animal, name):
    print(f"I have a {animal} named {name}.")

describe_pet("dog", "Buddy")
describe_pet("cat", "Whiskers")
describe_pet("parrot", "Polly")
````

````python
# 📌 Example 3: Function as a calculator
def add(a, b):
    print(f"{a} + {b} = {a + b}")

add(5, 3)
add(100, 200)
add(-10, 10)
````

````python
# 📌 Example 4: Building a profile
def build_profile(name, age, city):
    print(f"┌{'─'*30}┐")
    print(f"│ Name : {name:<22}│")
    print(f"│ Age  : {age:<22}│")
    print(f"│ City : {city:<22}│")
    print(f"└{'─'*30}┘")

build_profile("Alice", 25, "New York")
build_profile("Bob", 30, "London")
````

## 4. 🔄 The `return` Statement

The `return` statement sends a value **back** to the caller.  
Without `return`, a function returns `None` by default.

```
┌─────────────┐                    ┌─────────────┐
│   print()   │  → shows on screen │   return    │ → sends value back
│  (display)  │  → returns None    │  (output)   │ → can be stored
└─────────────┘                    └─────────────┘
```

````python
def say_hi():
  return "Hi! how are you?"
getting = say_hi()
print(getting)
````

````python
 #📌 Example 1: return vs print

# ❌ This just prints — we can't reuse the result
def add_print(a, b):
    print(a + b)

# ✅ This returns — we CAN reuse the result
def add_return(a, b):
    return a + b

# See the difference:
result1 = add_print(3, 4)     # Prints 7, but...
print(f"Stored value: {result1}")  # None!

print()

result2 = add_return(3, 4)    # Returns 7
````

````python
# 📌 Example 2: Using returned values in calculations
def square(n):
    return n ** 2

def cube(n):
    return n ** 3

x = 5
print(f"{x}² = {square(x)}")
print(f"{x}³ = {cube(x)}")
print(f"{x}² + {x}³ = {square(x) + cube(x)}")
````

````python
# 📌 Example 3: Returning multiple values (as a tuple)
def min_max(numbers):
    return min(numbers), max(numbers)

data = [45, 12, 89, 3, 67, 23]
lowest, highest = min_max(data)
# min_max(data)
print(f"Data: {data}")
print(f"Lowest : {lowest}")
print(f"Highest: {highest}")
````
