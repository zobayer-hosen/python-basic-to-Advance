## 4. 🔄 The `return` Statement

The `return` statement sends a value **back** to the caller.  
Without `return`, a function returns `None` by default.

```
┌─────────────┐                    ┌─────────────┐
│   print()   │  → shows on screen │   return    │ → sends value back
│  (display)  │  → returns None    │  (output)   │ → can be stored
└─────────────┘                    └─────────────┘
```

````python
def say_hi():
  return "Hi! how are you?"
getting = say_hi()
print(getting)
````

````python
 #📌 Example 1: return vs print

# ❌ This just prints — we can't reuse the result
def add_print(a, b):
    print(a + b)

# ✅ This returns — we CAN reuse the result
def add_return(a, b):
    return a + b

# See the difference:
result1 = add_print(3, 4)     # Prints 7, but...
print(f"Stored value: {result1}")  # None!

print()

result2 = add_return(3, 4)    # Returns 7
````

````python
# 📌 Example 2: Using returned values in calculations
def square(n):
    return n ** 2

def cube(n):
    return n ** 3

x = 5
print(f"{x}² = {square(x)}")
print(f"{x}³ = {cube(x)}")
print(f"{x}² + {x}³ = {square(x) + cube(x)}")
````

````python
# 📌 Example 3: Returning multiple values (as a tuple)
def min_max(numbers):
    return min(numbers), max(numbers)

data = [4,2,6,8,2,7,2]
lowest ,highest = min_max(data)

print(lowest)
print(highest)
print(data)
````

````python
#Example 4 : Returing a Dictionary
def get_circle_info(radius):
    import math
    return {
        "radius":radius,
        "diameter":2*radius,
        "circumference":round(2*math.pi* radius ,2),
        "area":round(math.pi *radius ** 2,2)
    }
info = get_circle_info(5)
print(info)
````

````python
# 📌 Example 5: Early return (guard clause)
def divide(a, b):
    if b == 0:
        return "❌ Error: Cannot divide by zero!"
    return round(a / b, 2)

print(divide(10, 3))
print(divide(10, 0))
print(divide(100, 7))
````

## 5. 🎯 Default Parameters

You can assign **default values** to parameters.  
If the caller doesn't provide a value, the default is used.

````python
def greet(name,getting = "hello"):
    print(f"{getting},{name}")
greet("Alice")
greet("Zobayer","good Morning")
greet("Charlie","Hey there")    
````

````python
# 📌 Example 2: Power function with default exponent
def power(base, exponent=2):
    return base ** exponent

print(f"5² = {power(5)}")        # Default: square
print(f"5³ = {power(5, 3)}")     # Override: cube
print(f"2¹⁰ = {power(2, 10)}")  # Override: 2^10
````

````python
# 📌 Example 3: Building a coffee order
def order_coffee(size="medium", drink="latte", milk="regular", extra_shot=False):
    shot = " + extra shot" if extra_shot else ""
    print(f"☕ Order: {size.title()} {drink}{shot} with {milk} milk")

order_coffee()                                          # All defaults
order_coffee("large")                                   # Override size
order_coffee("small", "cappuccino", "oat")             # Override 3
order_coffee("large", "avocado", extra_shot=True)  
````

## 6. 🔑 Keyword & Positional Arguments

| Type | How | Example |
|------|-----|---------|
| **Positional** | Order matters | `func(1, 2, 3)` |
| **Keyword** | Name matters | `func(x=1, y=2)` |

````python
# 📌 Example 1: Positional vs Keyword
def describe(name, age, city):
    print(f"{name} is {age} years old and lives in {city}.")

# Positional — order matters
describe("Alice", 25, "NYC")
# describe(25, "Alice", "NYC")
# Keyword — order doesn't matter!
describe(city="London", name="Bob", age=30)

# Mixed — positional first, then keyword
describe("Charlie", city="Tokyo", age=28)
````

## 7. 📦 `*args` and `**kwargs`

When you don't know how many arguments will be passed:

| Syntax | Collects | Type |
|--------|----------|------|
| `*args` | Extra positional args | `tuple` |
| `**kwargs` | Extra keyword args | `dict` |

````python
def add_all(*numbers):
    print(f"Received :{numbers} (type: {type(numbers).__name__})")
    return sum(numbers)

print(f" Sum = {add_all(1,2)}")
print(f" sum= {add_all(1,2,2,3,4,3)}")
print(f" Sum = {add_all(2,3,5,2,4,52)}")
````

````python
#Example 2 :** kwargs - Variable number of keyword arguments
def print_info(**details):
    print(f" Recived :{details} (type: {type(details).__name__})")
    for key, value in details.items():
        print(f" {key:>10}: {value}")
print_info(name="Alice", age=25)
print()
print_info(name="Bob", age=30, city="London", job="Engineer")        
````

## 8. 🔍 Scope — Local vs Global

```
┌───────────────── GLOBAL SCOPE ─────────────────┐
│  x = 10                                         │
│  ┌──────────── LOCAL SCOPE ──────────────┐      │
│  │  def my_func():                       │      │
│  │      y = 20  ← only exists here       │      │
│  │      print(x) ← CAN read global       │      │
│  └───────────────────────────────────────┘      │
│  print(y)  ← ❌ NameError!                      │
└─────────────────────────────────────────────────┘
```

````python
 #📌 Example 1: Helper function inside a main function
def process_grades(grades):
  def get_letter(score):
    if score>=90: return "A"
    elif score>=80: return "B"
    elif score>=70: return "C"
    else: return "F"

  print("  Score → Grade")
  print("  ─────────────")

  for score in grades:
    print(f"  {score:>5} → {get_letter(score)}")
process_grades([95, 82, 67, 55, 73, 91])    
````

## 12. 🎨 Decorators

A **decorator** wraps a function to extend its behavior without modifying it.  
They use the `@` syntax and are essentially closures that take a function as input.

```python
@decorator
def func():
    pass

# Is the same as:
func = decorator(func)
```
original function change na kore oi funtion ta ke wrapping kora,,,lets a  example of timing : kon function kottokon laglo seta mapa,,,logging: ke kon function call korse seta record raka ,, Auth: user Logged in kina check kora ,,, Caching: ager result mone raka ,bar bar calculation na kora

Decorator = "function-এর গায়ে extra কাপড় পরানো, কিন্তু মানুষটা একই থাকে।"

````python
def add_border(func):
    def wrapper(*args , **kwargs):
        print("="*30)
        func(*args, **kwargs)
        print("="*30)
    return wrapper 

@add_border
def say(message):
    print(f"||{message:<30}||")
say("hello decorator")    
  
````

````python
#time Decorator

import time
def timer(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print(f"{end-start:.2f}")
    return wrapper    

@timer
def slow_function():
    time.sleep(2)
    print("work is done")
slow_function()    
````

````python
def log_args(func):
    def wrapper(*args,**kwargs):
        print(f"calling {func.__name__}() with args:{args}")

        result = func(*args, **kwargs)
        print(f"Result: {result}")
        return result
    return wrapper

@log_args
def add(a,b):
    return a+b
result = add(3,4)
````

# Docstrings 

well-documented function are a sign of a professional developer

````python
def calculate_bmi(weight: float, height: float)-> dict:
    """Calculate body mass Index and return category .
     Args: 
     weight: Weight in Kilograms
     height: Height in meters

     Returns:
         A dictionary with 'bmi' value and 'catagory' string.
     Raises:
         ValueError: If weight or height is not positive.
     Ecamples:
         >>> calculate_bmi(70,1.75)
         {'bmi':22.86 ,'catagory':"Normal"}
     
     """
    if weight <= 0 or height <= 0:
        raise ValueError("Weight and height must be positive.")

    bmi = round(weight / height ** 2, 2)

    if bmi < 18.5:
        category = "Underweight"
    elif bmi < 25:
        category = "Normal"
    elif bmi < 30:
        category = "Overweight"
    else:
        category = "Obese"

    return {"bmi": bmi, "category": category}

# Test it
result = calculate_bmi(70, 1.75)
print(f"BMI: {result['bmi']} → {result['category']}")

# Access the docstring
print(f"\nDocstring:\n{calculate_bmi.__doc__}")
````
