Creating a LIST

```python
#empty
print([])
#1D
print([1,2,3,4,4])
#2D
print([1,2,3,[4,5]])
#3D
print([[[1,2],[3,4]],[[5,6],[7,8]]])
#Hetrogenous
print([1,True,5.6,5+6j,'hello'])
#using
print(list('hello'))
```

List Methods/Functions

list is mutable that means you can change the list value

```python
l= [13,4,5,6,7,8,4]
print(l)
l.append(5)
l.append(500)
print(l)
```

```python
L= [1,2,3,4]
L[2]= 200
print(L)
```

```python
#editing with slicing
L[1:4]=[200,3,3]
print(L)
```

```python
L.append([6,7,8])
print(L)
```

```python
L.append(True)
```

```python
L
```

Extend

```python
L= [1,2,3,4,5]
a = [10,11,12]
L.extend(a)
```

```python
L
```

Insert

```python
L= [1,2,3,4]
L.insert(1,200)
print(L)
```

```python
#deleting items from a list
L= [1,2,3,4]
del L[0]
print(L)
```

Remove

```python
L= [1,2,3,4,5]
L.remove(4)
print(L)

L.clear()
```

```python
L
```

Len/Min/Max/sorted

```python
L = [1,2,3,4]

print(len(L))
print(min(L))
print(max(L))
print(sorted(L,reverse=True))
```

Count

```python
L= [1,2,3,3,4,4,4,5,2]
L.count(4)
```
