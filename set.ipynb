#sets

A set is an unordered collection of items.every set elements is unique(no duplicaties) and must be immutable(canot be changed).
However, a set itself is mutable.we can add or remove items from it .
sets can do be used to perform mathematices set operations like union, interscetion, symmentric difference etc.

✅ List   → যেকোনো কিছু রাখা যায় (mutable/immutable)
✅ Dict   → value হিসেবে যেকোনো কিছু রাখা যায়
           কিন্তু key অবশ্যই immutable হতে হবে
❌ Set    → শুধু immutable রাখা যায় (hashable)

Characterstics:
Unorderd
mutable
no duplicates
cannot contain mutable data types

# set_methods
# len/sum/min/max/sorted

```python
s = {3,4,5,2,4,2}
print(len(s))
print(sum(s))
print(min(s))
print(max(s))
print(sorted(s,reverse=False))
```

```python
#union/update

s1= {1,2,3,4,5,6}
s2 = {4,5,2,6,33,5}

a= s1.union(s2)

print(a)
```

```python
s1.update(s2)
print(s1)
```

```python
# intersection/intersection_update
s1 = {1,2,3,4,5}
s2={2,3,4,5}
s1.intersection(s2)
```
