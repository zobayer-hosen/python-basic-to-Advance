# Dictionary

Dictionary in python is a collection of keys values, used to store data values like a map, which ,unlike other data types which hold only a single values as an element

In some language it is known as map or assosiative arrays:
dict = {"name":"zobayer", "age":22,"gender":"male"}
Characterstics:
mutable 
indexing has no meaning
keys can't be duplicated 
key can't be mutable items

# Create Dictionary

```python
# empty dictionary

d = {}
print(d)
print(type(d))
```

```python
d = set()
print(d)
```

```python
#1D dictionary
d1 = {"name":"zobayer","age":22,"gender":"male"}
print(d1)
```

```python
d2 = {[1,2,3,4]:333} # do no use list(mutable) as a key
print(d2)
```

```python
#2D dictionary
d3 = {
    "name":"bappy",
    "college_name":"ABC",
    "semester":5,
    "subject":{"DSA":80,"Math":50,"English":70}
}
print(d3)
```

```python
 
#using sequence and dict funtion 
d4 = dict([("name","zobayer"),("age",22)])
print(d4)
```

```python
#duplicate key
d5 = {"name":"bappy","name":"zobayer","age":22}
print(d5)
```
