Tuples
A tuple is a collection of values that are ordered and immutable. They are defined using parentheses () and can contain elements of different types.
In short ,a tuple is an immutable list.A tuple can not be changed in any way once it is created.

Tuple_Comprehension
you might read about list comprehension in python, but there is no tuple comprehension in python,but we will discuss how we can emulate tuple comprehension in python.

```python
#add i to 10 numbers to a list

l=[]
for i in range(1,11):
    l.append(i)
print(l)    
```
