```python
#custom LIST
class MyList:
    def __init__(self) -> None:
        self.data = []

    def append(self, item):
        self.data = self.data +[item] 
        return self.data
    def remove(self,item):
        if item in self.data:
            self.data = [x for x in self.data if x != item] 
            return self.data
        else:
            print(f"{item} list e nai") 

    def get(self,index):
        if 0<=index <len(self.data):
            return self.data[index]
        else:
            return "index ar bire"
        
    def length(self):
        return len(self.data)

    def show(self):
        print(f"List {self.data}")

    def clear(self):
        self.data = []
        print("the list is clear")

L = MyList()

L.append("zobayer")
L.append(1)
L.append(2)
L.remove(1)
L.get(1)
L.length()
L.show()
L.clear()
L.show()
            

```
